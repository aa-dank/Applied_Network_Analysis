```python
import networkx as nx
edges = [('a','c'), ('a', 'b'), ('b', 'd'), ('c', 'd'), ('c', 'e'), ('d', 'e'), ('d', 'g'), ('e', 'g'), ('g', 'f')]
G = nx.Graph()
G.add_edges_from(edges)
```

```python
%matplotlib notebook
nx.draw_networkx(G)
```

```python
nx.degree_centrality(G)
```

```python
nx.closeness_centrality(G)
```

```python
nx.betweenness_centrality(G, normalized = True, endpoints = False)
```

```python
nx.edge_betweenness_centrality(G, normalized = False)
```

```python
%matplotlib notebook
eds = [('a','b'), ('a', 'c'), ('c', 'a'), ('b', 'c'), ('d', 'c')]
Gd = nx.DiGraph()
Gd.add_edges_from(eds)

nx.draw_networkx(Gd)
```

```python
nx.hits(Gd, normalized = True)
```
